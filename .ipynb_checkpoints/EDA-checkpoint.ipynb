```python
# Import necessary libraries
import yfinance as yf
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from statsmodels.tsa.seasonal import seasonal_decompose
```

```python
# Set Seaborn style
sns.set_style("darkgrid")  
sns.set_context("talk")    
# Define the tickers
tickers = ['TSLA', 'BND', 'SPY']

# Load data
data = yf.download(tickers, start="2018-01-01", end="2023-01-01")['Adj Close']
print(data.head())  
```

```python
# Plotting
plt.figure(figsize=(12, 8))
data.plot(title="Stock Prices Over Time")
plt.xlabel("Date")
plt.ylabel("Adjusted Close Price")
plt.show()
```

```python
# Check basic information of the data
print(data.info())

# Check for missing values
print(data.isnull().sum())

# Handle missing values by forward filling, then backward filling any remaining NaNs
data.fillna(method='ffill', inplace=True)
data.fillna(method='bfill', inplace=True)

# Check basic statistics of the data
print(data.describe())

# Check data types
print(data.dtypes)
```

```python
# Normalize the data
data_normalized = (data - data.min()) / (data.max() - data.min())
data_normalized.head()
```

```python
# Plot closing prices over time
plt.figure(figsize=(14, 7))
for ticker in tickers:
    plt.plot(data.index, data[ticker], label=ticker)
plt.title("Adjusted Closing Prices Over Time")
plt.xlabel("Date")
plt.ylabel("Adjusted Closing Price")
plt.legend()
plt.show()
```

```python
# Calculate daily percentage change
returns = data.pct_change().dropna()

# Plot daily percentage change
plt.figure(figsize=(14, 7))
for ticker in tickers:
    plt.plot(returns.index, returns[ticker], label=f'{ticker} Daily % Change')
plt.title("Daily Percentage Change Over Time")
plt.xlabel("Date")
plt.ylabel("Daily Percentage Change")
plt.legend()
plt.show()
```

```python
# Calculate rolling statistics (e.g., 30-day rolling mean and standard deviation)
rolling_mean = data.rolling(window=30).mean()
rolling_std = data.rolling(window=30).std()

# Plot rolling mean and standard deviation for TSLA
plt.figure(figsize=(14, 7))
plt.plot(data['TSLA'], label='TSLA Closing Price', color='blue')
plt.plot(rolling_mean['TSLA'], label='30-Day Rolling Mean', color='orange')
plt.plot(rolling_std['TSLA'], label='30-Day Rolling Std Dev', color='green')
plt.title("TSLA Price with 30-Day Rolling Mean & Std Dev")
plt.xlabel("Date")
plt.ylabel("Price")
plt.legend()
plt.show()
```

```python
# Plot boxplot for outlier detection
plt.figure(figsize=(10, 5))
sns.boxplot(data=returns)
plt.title("Boxplot of Daily Returns for Outlier Detection")
plt.show()

# Identify days with unusually high or low returns for TSLA
threshold = 0.03  # Adjust threshold as needed
outliers = returns[(returns['TSLA'] > threshold) | (returns['TSLA'] < -threshold)]
outliers['TSLA']
```

```python
# Decompose the time series for TSLA to observe trend and seasonality
decompose_result = seasonal_decompose(data['TSLA'], model='multiplicative', period=365)
decompose_result.plot()
plt.show()
```

```python
# Calculate and plot 30-day rolling standard deviation to observe volatility trends for TSLA
tsla_volatility = returns['TSLA'].rolling(window=30).std()

plt.figure(figsize=(14, 7))
plt.plot(tsla_volatility, label='30-Day Rolling Volatility (TSLA)')
plt.title("30-Day Rolling Volatility for TSLA")
plt.xlabel("Date")
plt.ylabel("Volatility")
plt.legend()
plt.show()
```

```python
# Define parameters
confidence_level = 0.05  # 95% confidence level
risk_free_rate = 0.01    # Assume 1% annual risk-free rate

# Calculate VaR at 95% confidence level for TSLA
VaR_95 = returns['TSLA'].quantile(confidence_level)

# Calculate Sharpe Ratio for TSLA
mean_return = returns['TSLA'].mean() * 252  # Annualized return
std_dev = returns['TSLA'].std() * np.sqrt(252)  # Annualized standard deviation
sharpe_ratio = (mean_return - risk_free_rate) / std_dev

print(f"VaR at 95% confidence level for TSLA: {VaR_95}")
print(f"Sharpe Ratio for TSLA: {sharpe_ratio}")
```
